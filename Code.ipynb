```python
#Import Libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
# Load the dataset
df = pd.read_csv('indian_cities.csv')
df.head()
```

```python
# Function to recommend nearest locations
def recommend_locations(current_city, n_recommendations=5):
    # Check if the city exists in the dataset
    if current_city not in df['City'].values:
        return f"Error: '{current_city}' not found in the dataset. Please check the spelling."
   # Get the latitude and longitude of the current city
    user_location = df[df['City'] == current_city][['Latitude', 'Longitude']].iloc[0].values
    user_lat, user_lon = user_location
       # Extract latitude and longitude as a feature matrix
    locations = df[['Latitude', 'Longitude']].values
      # Initialize the Nearest Neighbors model
    knn = NearestNeighbors(n_neighbors=n_recommendations + 1, metric='euclidean')  # +1 to exclude the input city itself
    knn.fit(locations)

    # Find the nearest neighbors to the user's location
    distances, indices = knn.kneighbors([[user_lat, user_lon]])

    # Fetch the recommended locations, excluding the user's own city
    recommendations = df.iloc[indices[0][1:]].copy()  # Skip the first neighbor (self)
    recommendations['Distance (km)'] = distances[0][1:] * 111  # Approx. conversion to kilometers
    return recommendations

```

```python
# Take the user's current city as input
print("Enter your current city name:")
current_city = input("City: ").strip()

# Get nearest city recommendations
n_recommendations = 5  # You can adjust this number
recommended_cities = recommend_locations(current_city, n_recommendations)

# Display the recommendations
if isinstance(recommended_cities, str):  # If an error message is returned
    print(recommended_cities)
else:
    print("\nRecommended Locations:")
    print(recommended_cities[['City', 'State', 'Latitude', 'Longitude', 'Distance (km)']])
```
